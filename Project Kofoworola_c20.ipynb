## TASK C
    Compute student grades per subject (use the grading scheme below)
    

```python
 studentGrades = {}
```

```python
#GradingScheme
def computeGrade(score):
    if score >= 80:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 60:
        return 'C'
    elif score >= 50:
        return 'D'
    else:
        return 'F'


def computeStudGrades():
  
    for student in ChemStudents:
        index = ChemStudents.index(student)
        score = ChemScores[index]
        grade = computeGrade(score)
        studentGrades.setdefault(student, {})
        studentGrades[student]['Chemistry'] = grade
    
    for student in BioStudents:
        index = BioStudents.index(student)
        score = BioScores[index]
        grade = computeGrade(score)
        studentGrades.setdefault(student, {})
        studentGrades[student]['Biology'] = grade
    
    for student in PhyStudents:
        index = PhyStudents.index(student)
        score = PhyScores[index]
        grade = computeGrade(score)
        studentGrades.setdefault(student, {})
        studentGrades[student]['Physics'] = grade
    
    for student in MathStudents:
        index = MathStudents.index(student)
        score = MathScores[index]
        grade = computeGrade(score)
        studentGrades.setdefault(student, {})
        studentGrades[student]['Mathematics'] = grade
    
    return studentGrades
```

```python
grades = computeStudGrades()

# Print grades for each student
for student, subjects in grades.items():
    print(f"Grades for {student}:")
    for subject, grade in subjects.items():
        print(f"{subject}: {grade}")
    print()
```

## TASK D
    Compute final grade(All subject combined) for each student.

```python
#Below is the names and scores of each student
scores = {
    "David": [46,46,76,81], 
    "Samuel": [86,86,49],
    "Kelechi": [57, 57,69,63],
    "Chiamaka":[87, 88],
    "Oshinlolu":[98,55],
    "Belonwu": [45, 39, 99],
    "Samson": [78, 85, 77],
    "Chidiebere":[90, 81, 78],
    "Mercy": [84, 90],
    "Efe": [67, 70, 78],
    "Bello": [54, 56, 60],
    "Maryrose": [59, 68, 54]
}
```

```python
finalGrades = {}
```

```python
def compStudFinalGrades():
    for name, scoreList in scores.items():
        aveScore = sum(scoreList) / len(scoreList)
        finalGrade = compFinalGrade(aveScore)
        finalGrades[name] = finalGrade

    return finalGrades


def compFinalGrade(score):
    if score >= 80:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 60:
        return 'C'
    elif score >= 50:
        return 'D'
    else:
        return 'F'
```

```python
finalGrades = compStudFinalGrades()

for name, grade in finalGrades.items():
    print(f"Final Grade for {name}: {grade}")
```

## TASK E
    Assign Exam Status:
    a. If student has all A, B or C in all subjects --> Pass
    b. If student has atleast a D in any one subject --> Retake Exam (Subjects)
    c. If student has an F in atleast one subject --> Fail(Subjects)

```python
 examStatus = {}
```

```python
def assignExamStatus(studentGrades):
    for student, subjects in studentGrades.items():
        grades = list(subjects.values())

        if all(grade in ['A', 'B', 'C'] for grade in grades):
            examStatus[student] = 'Pass Examination'
        elif 'D' in grades:
            examStatus[student] = 'Retake Examination (Subjects)'
        else:
            examStatus[student] = 'Fail (Subjects)'

    return examStatus
```

```python
examStatus = assignExamStatus(grades)

# Print exam status for each student
for student, status in examStatus.items():
    print(f"Exam status for {student}: {status}")
```

## TASK G
    Generate a table of exam results for each subject

```python
subjects = set()
subjectGrades = {}
```

```python
def ExamResultsTable():
    # Gather unique subjects and their grades
    for grades in studentGrades.values():
        for subject, grade in grades.items():
            if subject != "Overall Grade":
                subjects.add(subject)
                if subject not in subjectGrades:
                    subjectGrades[subject] = []
                subjectGrades[subject].append(grade)

    # Create a DataFrame for the exam results table
    dataTable = pd.DataFrame(subjectGrades)
    dataTable.index.name = "Student"
    
    # Display the exam results table
    display(dataTable)
```

```python
# Generate the exam results table
ExamResultsTable()
```
